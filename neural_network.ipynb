```python
def sigmoid(X):
    return 1/(1+np.exp(-X))
```

```python
def neuron(X,w1,b1):
    z=np.dot(X,w1)+b1;
    _pred=sigmoid(z);
    print(_pred)
    y_pred=np.round(_pred);
    return y_pred
```

```python
output=neuron(X,w1,b1)
print(output)
```

```python
def sigmoid_derivative(X):
    return X*(1-X)
```

```python
lr = 0.05
epochs = 50000
n_samples = 4
```

```python
for epoch in range(epochs):
    z1 = np.dot(X, w1) + b1         
    a1 = sigmoid(z1)
    
    #z2 = np.dot(a1, w2) + b2       
    #a2 = sigmoid(z2)
    
    loss = np.mean((y - a1)**2)
    error = -2 * (y - a1) / n_samples  
    
    d2 = error * sigmoid_derivative(a1)     
    #d1 = np.dot(d2, w2.T) * sigmoid_deriv(a1)  
    
    #w2 -= lr * np.dot(a1.T, d2)        
    #b2 -= lr * np.sum(d2, axis=0)      
    w1 -= lr * np.dot(np.array(X).T, d2)   
    b1 -= lr * np.sum(d2, axis=0)  # Shape becomes (1,)
  
    
    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")

print("Final Output:\n", np.round(a1))
```
